## Exercice 'cadre'

```python
print(np.__version__)
```

```python
square = np.array([[True]*10]*10)
```

```python
plt.imshow(square) ;
```

```python
def cadre(tab, inplace=True):
    assert np.ndim(tab) == 2
    if inplace:
        tab[1,:] = False
        tab[-2,:] = False
        tab[:,1] = False
        tab[:,-2] = False
        return tab
    else:
        arr = np.copy(tab)
        return cadre(arr)
        pass
        
```

```python
plt.imshow(cadre(square,False)) ;
```

```python
rect = np.array([[0]*8]*6, dtype=np.uint8)
for i in range(6):
    for j in range(8):
        rect[i,j] = np.random.randint(0,21)
        
plt.imshow(rect) ;
```

```python
plt.imshow(cadre(rect,False))
```

# Exercice 'cadre 2'

```python
def cadre2(tab):
    (l,c) = np.shape(tab)
    assert np.ndim(tab) == 2
    arr = np.insert(tab,[0,c],False, axis = 1)
    arr = np.transpose(arr)
    arr = np.insert(arr, [0,l],False, axis = 1)
    return np.transpose(arr)
```

```python
plt.imshow(cadre2(square)) ;
```

```python
np.shape(cadre2(square))
```

```python
plt.imshow(cadre2(rect)) ;
```
