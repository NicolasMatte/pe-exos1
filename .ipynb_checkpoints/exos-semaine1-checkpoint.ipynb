## Exo cadre

```python
print(np.__version__)
```

```python
square = np.array([[True]*10]*10)
```

```python
plt.imshow(square) ;
```

```python
def cadre(tab, inplace=True):
    assert np.ndim(tab) == 2
    if inplace:
        tab[1,:] = False
        tab[-2,:] = False
        tab[:,1] = False
        tab[:,-2] = False
        return tab
    else:
        array = np.copy(tab)
        return cadre(array)
        pass
        
```

```python
plt.imshow(cadre(square,False)) ;
```

```python
rect = np.array([[0]*8]*6, dtype=np.uint8)
for i in range(6):
    for j in range(8):
        rect[i,j] = np.random.randint(0,21)
        
plt.imshow(rect)
```

```python
plt.imshow(cadre(rect))
```

```python
On constate que la fonction cadre remplit bien sa fonction.
```
