# Exercice 'cadre 2'

```python
def cadre2(tab):
    (l,c) = np.shape(tab)
    assert np.ndim(tab) == 2
    arr = np.insert(tab,[0,c],False, axis = 1)
    arr = np.transpose(arr)
    arr = np.insert(arr, [0,l],False, axis = 1)
    return np.transpose(arr)
```

```python
plt.imshow(cadre2(square)) ;
```

```python
np.shape(cadre2(square))
```

```python
plt.imshow(cadre2(rect)) ;
```

## Exercice 'anti_diagonal'

```python
def anti_diagonal(liste):
    n = len(liste)
    mat = np.zeros((n,n),float)
    for i in range(n):
        mat[i,n-1-i] = liste[i]
    return mat
```

```python
anti_diagonal([1])
```

```python
anti_diagonal([1,2])
```

```python
anti_diagonal([1,2,3,4])
```

```python
def anti_diagonal1(liste):
    n = len(liste)
    mat = np.empty((n,n),float)
    for i in range(n):
        mat[i,n-1-i] = liste[i]
    return mat
```

```python
anti_diagonal1([1])
```

```python
anti_diagonal1([1,2])
```

```python
anti_diagonal1([1,2,3,4])
```

```python
np.empty((3,2))
```

```python
def anti_diagonal2(liste, dtype = float):
    mat = np.fliplr(np.diag(liste))
    return np.array(mat, dtype)
```

```python
anti_diagonal2([True,False], bool)
```

```python
anti_diagonal2([True,False])
```

```python
anti_diagonal2([2,4,8], np.uint8)
```

```python
anti_diagonal2((1,2,3),np.uint8)
```

```python
anti_diagonal2({1,2,3},np.uint8)
```

```python
anti_diagonal2({0:'a',1:'b',2:'c'})
```

```python
anti_diagonal2(range(4), dtype=np.int16)
```

```python
anti_diagonal2(x**2 for x in range(4))
```

```python
anti_diagonal([x**2 for x in range(4)])
```
